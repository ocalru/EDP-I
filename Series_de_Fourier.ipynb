 Definimos $f_1(x) = 0 $ como la primera parte de la funcion a trozos:

 $$
f(x) =
\begin{cases}
0, & \text{si } -2 ≤ x ≤ 0 \\
-x, & \text{si } 0 ≤ x ≤ 2
\end{cases}
$$

```python
def recta(x): # se define la funcion "recta" en x
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 0
    return y
```

```python
x = linspace(-2,0,100)
f = recta(x)
plot(x,f,label='$f_1(x) = 0$')
plt.xlabel('$x$'); plt.ylabel('$y$');
plt.title("$f_1(x) = 0$")
xlim([-2.01,0.01]),ylim([-1.01,1.01]),grid(),legend()
show()
```

Usamos una serie de Fourier para aproximar $f_1(x) = 0$




```python
def f_1(x,N=1000): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (8)*((-1)**n)*sin(n*pi*x)/(n**2*pi)
    return (-0.5)*f
```

```python
f_n = f_1(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.01,0.01]),ylim([-1.01,1.01]),grid(),legend()
show()
```

 Definimos $f_2(x) = -x $ como la segunda parte de la funcion a trozos:

 $$
f(x) =
\begin{cases}
0, & \text{si } -2 ≤ x ≤ 0 \\
-x, & \text{si } 0 ≤ x ≤ 2
\end{cases}
$$

```python
def recta1(x):  # Se define la funcion "recta" en x
    y = -x  # Simplemente calculamos y = -x
    return y

x = linspace(0, 2, 1000)
y = recta1(x)

plot(x, y, label='$f_1(x) = -x$')
xlabel('$x$')
ylabel('$y$')
title("$f_1(x) = -x$")
xlim([1000, 1000])
ylim([1000, 10000])
grid()
legend()
show()
```
